```python
import matplotlib.pyplot as plt
from maad import sound, util

# variables
nperseg = 1025  # window size for STFT
noverlap = 512  # window overlap for STFT
db_range = 90   # dB range, clip background noise
cmap = 'gray'   # colormap

s, fs = sound.load_url('spinetail')
Sxx, _, _, ext = sound.spectrogram(s, fs, nperseg=nperseg, noverlap=noverlap)
fig, ax = plt.subplots(figsize=(15, 5))
util.plot_spectrogram(Sxx, ext, db_range=db_range, ax=ax, colorbar=False, cmap=cmap)
ax.set_ylabel('Frequency (kHz)')
ax.set_xlabel('Time (s)')
```
